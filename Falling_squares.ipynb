```python
class Solution:
    def fallingSquares(self, positions: List[List[int]]) -> List[int]:
        positions = np.array(positions)
        starting = positions[:, 0]
        ending = positions[:, 0] + positions[:, 1]
        all_positions = np.concatenate((starting, ending))
        all_positions.sort()

        all_positions = np.unique(all_positions).reshape(-1, 1)

        height = np.zeros((all_positions.shape[0], 1))
        height = np.concatenate((all_positions, height), axis=1)

        ans = []

        for i in range(positions.shape[0]):
            start = positions[i, 0]
            square_side = positions[i, 1]
            end = start + square_side
            # calculated the "affected" rows in the array - i.e., positions along
            # the x-axis which whose height will change because of dropping this square
            affected = ((height[:, 0] >= start) * (height[:, 0] < end))
            print(i, affected)
            curr_peak = height[affected, 1].max()
            height[affected, 1] = curr_peak + square_side
            ans.append(int(max(height[:, 1])))
        print(height)
        return ans
```
